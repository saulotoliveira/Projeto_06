## 4. Verificar pacotes instalados

```python
import importlib

pacotes_principais = [
    "numpy", "pandas", "matplotlib", "seaborn",
    "scikit-learn", "scipy", "statsmodels", "requests"
]

print("Verificação dos pacotes principais:")
print("-" * 40)
for pacote in pacotes_principais:
    nome_import = pacote.replace("-", "_")
    try:
        mod = importlib.import_module(nome_import)
        versao = getattr(mod, "__version__", "N/A")
        print(f"  {pacote}: OK (v{versao})")
    except ImportError:
        print(f"  {pacote}: FALHOU")

print("-" * 40)
print("Instalação concluída!")
```
